## setting up IDE to use the virtual enviornment

Since we set up all the required packages in a conda virtual enviornment, we need to explicitly tell the IDE or whatever we are running to use that enviornment, instead of the system default.

Now it's perfect time to explain what I mean by "using that enviornment", and further more, how all these python terminals and virtual enviornments actually works.

Simply put, a terminal is a purly text-based interface to all kinds of applications. In __[Unix-like systems](https://en.wikipedia.org/wiki/Unix-like)__, a lot of, if not all the stuff that can be done through graphical UI (i.e. clicking), can also be done through text-based commands, with the help of __[some sort of terminals](https://en.wikipedia.org/wiki/Unix_shell)__. So how does the commands work? They are basicly little applications that get called everytime you issued a command. For example, in the CaImAn install process, this line:
`git clone https://github.com/simonsfoundation/CaImAn`
calls for a little program named `git`, the rest of the line is passed to the `git` to tell it what to do. You can simply type in `git` without telling it anything, for most of command-line applications this will result in printing out a help page that tells you what the application can do. You can actually find the `clone` subcommands in this help page (ignore the prepending "!", you should not add it in your terminal):

````python
!git
````

Of course it's the same for `conda` too:

````python
!conda
````

As you can imagine, there are millions of little programs, and everytime you install something new, a (or some) new command will become available. How does the terminal magically find the application you want with only a single word? It's __[through the idea of `$PATH`](https://askubuntu.com/questions/482971/how-does-the-terminal-work)__, which is basically a collection of folders where all those little applications rests. Everytime you issue a command, the terminal search in your `$PATH` (those folders), and calls for any program whose name matches your input. Of course, it will complain if nothing is found, for example:

````python
!blah_blah_blah_something_that_dont_exsist
````

So, unlike Windows, where installing and uninstalling programs is a state of magic, in UNIX-like systems, installing a program is simply copying the programs (or codes, essentially) to the appropriate `$PATH`, since in UNIX-like systems __[everything is a file](https://en.wikipedia.org/wiki/Everything_is_a_file)__ and there's nothing special about programs.

That being said, you can imagine a circumstance where you have different version of the same program (with the same name) under different folders, and select which one you want to use by setting the search `$PATH` everytime you call for that program. This sounds like nonsense for something like office, but is extremely useful for something like python, since python has two major versions (2.x and 3.x), and all those packages depending on python will differ in some way and conflict with each other if messed up together. This is exactly what conda helped you do - create sub-folders, put python and desired additional packages in them and manage them for you, and each of those folders can be considered a self-contained virtual enviornment.

Let's see these for ourselves. By default, the directory of each virtual enviornment is simply __ROOT_DIR/envs/the_name_of_your_enviornment__ where __[ROOT_DIR is the directory of your conda installation](https://conda.io/docs/user-guide/concepts.html)__. Again you can easily find these information by __[listing your enviornments](https://conda.io/docs/user-guide/tasks/manage-environments.html#viewing-a-list-of-your-environments)__.

````python
!conda env list
````

You can see my current activated enviornment is __CaImAn__ (signified by a star). Here is what is looks like in the enviornment directory:

````python
!tree -d -L 2 /opt/miniconda3/envs/CaImAn
````

The python executable file specific to this enviornment is actually in the __bin__ folder. We can directly invoke python there:

````python
!/opt/miniconda3/envs/CaImAn/bin/python -c 'print("hey there!")'
````

That said, each time you do `source activate some_enviornment`, what's actually happening is that conda point the `$PATH` variable to the desired enviornment directory for you, so that afterwards when you call `python`, it's actually running the python executable file inside that enviornment folder, instead of the system default one.

Now, it turns out that python use exactly the same idea to search for all those additional packages - a `$PATH` variable. The difference is that python has its own __[`$PYTHONPATH`](https://docs.python.org/2/using/cmdline.html#envvar-PYTHONPATH)__ variable which is distinct from the global system-default `$PATH` variable. Whenever you ask for a module (i.e. packages) in python, it search in the directories listed under `$PYTHONPATH`. Conveniently, when you invoke `python` from different places, the `$PYTHONPATH` variable is __[automatically set relative to the python executable](https://leemendelowitz.github.io/blog/how-does-python-find-packages.html)__, so neither you nor conda need to worry about setting that properly. To demonstrate, let's invoke python from different places and ask it to tell us which `$PYTHONPATH` it's using (the string after `-c` is a tiny piece of python code printing out the `$PYTHONPATH` variable, you don't have to worry about that):

````python
!/opt/miniconda3/envs/CaImAn/bin/python -c 'import sys; print("\n".join(sys.path))'
````

As you can see, the python in __CaImAn__ enviornment is searching for directories under __/opt/miniconda3/envs/CaImAn/lib__, which will change if we call for another python from difference place:

````python
!/opt/miniconda3/envs/calblitz/bin/python -c 'import sys; print("\n".join(sys.path))'
````

This is a long explanation about how the conda enviornments work and hopefully it make sense. Now we can set out and ask our IDE to use the enviornments. In most cases, it's very simple - make sure you `source activate your_enviornment`, and then simply invoke `python`, `ipython` or `spyder` from there (hopefully you already figured this out yourself after the long explanation). For more complicated IDEs who like to control everything inside their program (like PyCharm), this does not work. But the basic idea stays the same - you need to point the IDE to the proper python executable you want to use and then you're set. __[Here](https://www.jetbrains.com/help/pycharm/configuring-python-interpreter.html)__ is the official guide of how to do that in PyCharm (follow the "Configuring a local interpreter" part - we don't need to use the `virtualenv` tool from PyCharm since we already have conda doing that for us). If you are using neither spyder or PyCharm, the best way to find out how to do this is __[google "[your IDE name] + set python interpreter"](https://www.google.com/search?client=ubuntu&hs=0FY&channel=fs&q=pycharm+set+python+interpreter&spell=1&sa=X&ved=0ahUKEwjXuoaJvvrWAhUGJiYKHZ6FANEQvwUIIygA&biw=1440&bih=793)__.
